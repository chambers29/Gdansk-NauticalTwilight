```python
import requests
import pandas as pd
from datetime import datetime, timedelta

lat, lng = 54.262203, 18.636150  # PG
start_date = datetime(2025, 1, 1)
end_date = datetime(2025, 12, 31)

data = []

for day in range((end_date - start_date).days + 1):
    date = (start_date + timedelta(days=day)).strftime("%Y-%m-%d")
    print("Pobieram dane dla:", date)
    url = f"https://api.sunrise-sunset.org/json?lat={lat}&lng={lng}&date={date}&formatted=0"
    response = requests.get(url).json()
    results = response['results']
    
    data.append({
        "date": date,
        "sunrise": results['sunrise'],
        "sunset": results['sunset'],
        "nautical_twilight_begin": results['nautical_twilight_begin'],
        "civil_twilight_begin": results['civil_twilight_begin'],
        "day_length": results['day_length']
    })

df = pd.DataFrame(data)
df['date'] = pd.to_datetime(df['date'])
df.to_csv("sunrise_2025.csv", index=False)
print(df.head())
```
